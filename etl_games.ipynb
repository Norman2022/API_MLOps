```python
import json
import pandas as pd
import ast
```

```python
data = []
with open('data/output_steam_games.json') as f:
    for line in f:
        data.append(json.loads(line))
```

```python
df_games = pd.DataFrame(data)
```

```python
df_games_copy = df_games.copy()
```

```python
#vemos que poseen datos similares Entonces completamos los fatantes en uno con el otro
df_games_copy['app_name'].fillna(df_games_copy['title'], inplace=True)
df_games_copy['title'].fillna(df_games_copy['app_name'], inplace=True)
```

```python
#verifico las filas con muchos NaN
df_games_copy[df_games_copy.isnull().sum(axis=1)>=15]
```

```python
#Filtramos las que contienen NaN distinto al analisis anterior
df_games_copy = df_games_copy[df_games_copy.isnull().sum(axis=1)!=15]
df_games_copy.reset_index(drop=True, inplace=True) 
df_games_copy.info()
```

```python
df_games_copy.isnull().sum()
```

```python
#vemos si podemos salvar algunos de las filas con id NaN
df_games_copy[df_games_copy.id.isnull()]
```

```python
#completamos el id con steamId
df_games_copy.id.iloc[30960] = 200260
df_games_copy.reset_index(drop=True, inplace=True) 
```

```python
#vemos si podemos salvar algunos titulos, cuyas filas son NaN
df_games_copy[df_games_copy.title.isnull()]
```

```python
df_games_copy = df_games_copy.drop(df_games_copy.index[2579])
df_games_copy.reset_index(drop=True, inplace=True) 
```

```python
df_games_copy.isnull().sum()
```

```python
#vemos que poseen datos similares Entonces completamos los fatantes en uno con el otro
df_games_copy['genres'].fillna(df_games_copy['tags'], inplace=True)
df_games_copy['tags'].fillna(df_games_copy['genres'], inplace=True)
```

```python
df_games_copy = df_games_copy.drop('tags', axis = 1)
```

```python
df_games_copy.isnull().sum()
```

```python
#funcion para convertir palabras en precio = 0
def convertir_precio(valor):
    try:
        return float(valor)
    except (ValueError, TypeError):
        return 0
```

```python
df_games_copy['price'] = df_games_copy['price'].apply(convertir_precio).fillna(0)
```

```python
df_games_copy.isnull().sum()
```

```python
df_games_copy = df_games_copy.dropna(subset=['genres','release_date'])
```

```python
df_games_copy.isnull().sum()
```

```python
df_games_copy.info()
```

```python
df_games_copy
```

```python
#eliminamos columnas con mucha informacion faltante: metascore, discount_price;  
#app_name por ser similar a title, y las url por no aportar valor
df_games_copy.drop(['app_name','metascore','discount_price','reviews_url','url', 'publisher', 'developer' ], inplace=True, axis=1)
df_games_copy.drop(['specs','user_id','steam_id','items','items_count',  'early_access', 'price' ], inplace=True, axis=1)
```

```python
#cambiamos el tipo de datos de id para poder gurdar como parquet sin que de error y renombramos
df_games_copy['id'] = df_games_copy['id'].apply(lambda x: x.to_bytes(4, byteorder='big') if isinstance(x, int) else x)
df_games_copy.rename(columns={'id': 'item_id'}, inplace=True)
df_games_copy.reset_index(drop=True, inplace=True)
```

```python
df_games = pd.read_parquet('data_games.parquet')
```

```python
df_games['item_id'] = df_games['item_id'].str.decode('utf-8')
# Convertir la columna 'release_date' a tipo datetime, manejando errores
df_games['release_date'] = pd.to_datetime(df_games['release_date'], errors='coerce')
# Filtrar filas con fechas válidas (no NaN)
df_games = df_games.dropna(subset=['release_date'])
#expandir la columna genres
df_games = df_games.explode('genres',ignore_index=True)
#filtramos y llevamos a minusculas
df_games['genres'] = df_games['genres'].fillna('')
df_games['genres'] = df_games['genres'].str.lower()
```

```python
df_games.drop('price',axis =1, inplace=True)
```

```python
nombre_archivo_parquet = "data_games_explode.parquet"
# Guarda el DataFrame como un archivo Parquet
df_games.to_parquet(nombre_archivo_parquet)
```
